# Gaussian Blur for image smoothing

```python
import cv2
import matplotlib.pyplot as plt
```

```python
path_image = '../data/raw/hand_xray.jpeg'
image = cv2.imread(path_image, cv2.IMREAD_GRAYSCALE)
```

```python
#Aplly Gaussian Blur
blurred_image = cv2.GaussianBlur(image, ksize= (35, 35), sigmaX= 0)
#ksize = Tamanho da lente para desfoque. Quanto maior mais borrado (√çMPAR)
#sigmaX = Grau de desfoque. Deixar em 0 faz com que o OpenCV calcule para gente
```

```python
plt.Figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original')

plt.subplot(1, 2, 2)
plt.imshow(blurred_image, cmap='gray')
plt.title('Imagem desfocada')

plt.show()
```
