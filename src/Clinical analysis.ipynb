# Tests for clinic cases

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
path_image = '../data/raw/patella-fracture-6.jpg'
image = cv2.imread(path_image, cv2.IMREAD_GRAYSCALE)
```

Sharpering

```python
kernel_shapering = np.array([
    [0, -1, 0],
    [-1, 5, -1],
    [0, -1, 0]
])
```

```python
sharpened_image = cv2.filter2D(image, -1, kernel_shapering)
```

```python
plt.figure(figsize=(18, 12))

plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original')

plt.subplot(1, 2, 2)
plt.imshow(sharpened_image, cmap='gray')
plt.title('Sharpened Image')

plt.show()
```

INVERTING AND APPLY CLAHE

```python
inverted_image = 255 - sharpened_image
```

```python
clahe = cv2.createCLAHE(clipLimit=2.0 , tileGridSize=(8, 8))
inverted_clahe = clahe.apply(inverted_image)
```

```python
plt.figure(figsize=(18, 12))

plt.subplot(1, 2, 1)
plt.imshow(sharpened_image, cmap='gray')
plt.title('Sharpened Image')

plt.subplot(1, 2, 2)
plt.imshow(inverted_clahe, cmap='gray')
plt.title('Inverted Sharpened Image with CLAHE')

plt.show()
```

Blurred image

```python
blurred_image = cv2.GaussianBlur(inverted_clahe, (5, 5), 0)
```

# Final result

```python
plt.figure(figsize=(18, 12))

plt.imshow(blurred_image, cmap='gray')
plt.title('Canny in a inverted clahe image')

plt.show()
```
