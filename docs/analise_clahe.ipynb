# Image analysis: Comparing CLAHE in Original Image vs. Inverted

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
path_image = '../data/raw/raio_x_chest.png'
image = cv2.imread(path_image, cv2.IMREAD_GRAYSCALE)

inverted_image = 255 - image

clahe = cv2.createCLAHE(clipLimit = 2.0, tileGridSize=(8, 8))

clahe_image = clahe.apply(image)
clahe_inverted = clahe.apply(inverted_image)
```

```python
plt.figure(figsize=(12, 10))

plt.subplot(1, 4, 1)
plt.imshow(image, cmap='gray')
plt.title('Original')

plt.subplot(1, 4, 2)
plt.imshow(inverted_image, cmap='gray')
plt.title('Inverted')

plt.subplot(1, 4, 3)
plt.imshow(clahe_image, cmap='gray')
plt.title('Original clahe')

plt.subplot(1, 4, 4)
plt.imshow(clahe_inverted, cmap='gray')
plt.title('Inverted clahe')

plt.tight_layout()
plt.show()
```
