# fracture analysis

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
path_image = '../../data/raw/patella-fracture-6.jpg'
image = cv2.imread(path_image, cv2.IMREAD_GRAYSCALE)
```

THIS PROJECT CONSISTS OF A ANALYSIS OF PIPELINES OF PRE-PROCESSING IN FRACTURES

```python
# RECIPE 1
# ONLY SHARPENING
kernel_shapering = np.array([
    [0, -1, 0],
    [-1, 5, -1],
    [0, -1, 0]
])
sharpened_image = cv2.filter2D(image, -1, kernel_shapering)

plt.figure(figsize=(18, 12))

plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('ORIGINAL')

plt.subplot(1, 2, 2)
plt.imshow(sharpened_image, cmap='gray')
plt.title('Sharpened image')

plt.show()
```

```python
# RECIPE 2
# INVERTED AND CLAHE
inverted_image = 255 - image

clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8, 8))
image_clahe = clahe.apply(inverted_image)

plt.figure(figsize=(18, 12))

plt.subplot(1, 2, 1)
plt.imshow(sharpened_image, cmap='gray')
plt.title('Sharpened image')

plt.subplot(1, 2, 2)
plt.imshow(image_clahe, cmap='gray')
plt.title('Inverted clahe image')

plt.show()
```

```python
# RECIPE 2.1
# CLAHE -> INVERTED
image_clahe_2 = clahe.apply(image)

image_claheinverted = 255 - image_clahe_2

plt.figure(figsize=(18, 12))

plt.subplot(1, 2, 1)
plt.imshow(image_clahe, cmap='gray')
plt.title('Inverted clahe image')

plt.subplot(1, 2, 2)
plt.imshow(image_claheinverted, cmap='gray')
plt.title('Clahe inverted image')

plt.show()
```
