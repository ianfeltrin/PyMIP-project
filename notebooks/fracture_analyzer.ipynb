# Origial vs CANNY

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
path_image = '../data/raw/patella-fracture-6.jpg'
image = cv2.imread(path_image, cv2.IMREAD_GRAYSCALE)
```

```python
#suavizar
#blurred_image = cv2.GaussianBlur(image, (9,9), 0)
blurred_image = cv2.medianBlur(image, ksize= 5)
canny_image = cv2.Canny(blurred_image, 50, 50)
```

```python
plt.figure(figsize=(18, 12))

plt.subplot(2, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original')

plt.subplot(2, 2, 2)
plt.imshow(canny_image, cmap='gray')
plt.title('Canny image')

plt.subplot(2, 2, 3)
plt.imshow(blurred_image, cmap='gray')

plt.tight_layout()
plt.show()
```

 ADICIONANDO CLAHE

```python
clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8, 8))
image_clahe = clahe.apply(image)
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(2, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original')

plt.subplot(2, 2, 2)
plt.imshow(image_clahe, cmap='gray')
plt.title('Clahe image')

plt.subplot(2, 2, 3)
plt.imshow(canny_image, cmap='gray')
plt.title('Canny image')

plt.tight_layout()
plt.show()
```

UTILIZADO A TCNICA DE CLAHE COM INVERSO

```python
image_inverted = 255 - image
clahe_inverted = clahe.apply(image_inverted)
```

```python
plt.figure(figsize=(12, 12))

plt.subplot(2, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original')

plt.subplot(2, 2, 2)
plt.imshow(image_clahe, cmap='gray')
plt.title('Clahe image')

plt.subplot(2, 2, 3)
plt.imshow(canny_image, cmap='gray')
plt.title('Canny image')

plt.subplot(2, 2, 4)
plt.imshow(clahe_inverted, cmap='gray')
plt.title('Inverted clahe')

plt.tight_layout()
plt.savefig('../data/processed/fracture_analyzer_panel.png')

plt.show()
```
